1. What is the difference between the following two linee? Which one will give a larger value? Will this be the case for other texts?
* sorted(set([w.lower() for w in text1]))
* sorted([w.lower() for w in set(text1)])

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from nltk.book import *
```

```python
opt1 = sorted(set([w.lower() for w in text1]))
print(opt1)
print(len(opt1))
```

```python
opt2 = sorted([w.lower() for w in set(text1)])
print(opt2)
print(len(opt2))
```

The difference in these two lines of code is the order of operations. The first line of code will convert all the lowercase first and then remove duplicates. The second will first remove duplicates and then convert to lowercase. Since line two removes duplicates BEFORE transforming to lowercase, there will end up being duplicates left over in the word list making this the larger of the two.

3. Find all four-letter words in the Chat Corpus (text5). With the help of a frequency distribution (FreqDist), show these words in decreasing order of frequency.

```python
fours = [w for w in set(text5) if len(w) == 4]
fdist1 = FreqDist(fours)
for word, frequency in fdist1.most_common():
    print(f'{word}: {frequency}')
```

11. Define a function called vocab_size(text) that has a single parameter for the text, and which returns the vocabulary size of the text. 


```python
def vocab_size(text):
    unique_words = set(word.lower() for word in text)
    return len(unique_words)
```

12. Define a function percent(word, text) that calculates how often a given word occurs in a text and expresses the result as a percentage. 

```python
def percent(word, text):
    total_words = len(text)
    word_count = text.count(word.lower())
    percentage = (word_count/total_words)* 100 if total_words > 0 else 0
    return percentage
```
